```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
truncation_length_list = np.arange(100, 501, 100)
extraction_accuracy_list = np.zeros_like(truncation_length_list, dtype=float)
```

```python
for idx, truncation_length in enumerate(truncation_length_list):
    # Load the data from json files
    file_name = f"truncated_answer_results/maxlen_{truncation_length}/extracted_answers.json"
    df = pd.read_json(file_name)
    accuracy = df['extraction_accuracy'][0]
    print(f"accuracy: {accuracy}")
    # Store the accuracy for each truncation length
    extraction_accuracy_list[idx] = accuracy
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(truncation_length_list, extraction_accuracy_list, marker='o')
plt.title("Extraction Accuracy vs. Truncation Length")
```
